## Ejercicio

Escribir un programa que permita gestionar la base de datos de clientes de una empresa. Los clientes se guardarán en un diccionario en el que la clave de cada cliente será su NIF, y el valor será otro diccionario con los datos del cliente (nombre, dirección, teléfono, correo, preferente), donde preferente tendrá el valor `True` si se trata de un cliente preferente. El programa debe preguntar al usuario por una opción del siguiente menú: (1) Añadir cliente, (2) Eliminar cliente, (3) Mostrar cliente, (4) Listar todos los clientes, (5) Listar clientes preferentes, (6) Terminar. En función de la opción elegida el programa tendrá que hacer lo siguiente:

1. Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos.
2. Preguntar por el NIF del cliente y eliminar sus datos de la base de datos.
3. Preguntar por el NIF del cliente y mostrar sus datos.
4. Mostrar lista de todos los clientes de la base datos con su NIF y nombre.
5. Mostrar la lista de clientes preferentes de la base de datos con su NIF y nombre.
6. Terminar el programa.

## Solución

```python
clientes = {
    27408469739 : {
        "nombre" : "Juan",
        "direccion" : "Las Heras",
        "telefono" : "2615089367",
        "correo" : "juan@gmail.com",
        "preferente" : False
    }
}

def mostrar_clientes():
    print("\nListado actual de clientes:")
    for cuil, datos in clientes.items():
        for dato, valor in datos.items():
            print(dato,":",valor, end=" , ")
        print("CUIL :",cuil)
        
mostrar_clientes()

def añadir_cliente():
    condicion = True
    while condicion == True:
        datos_cliente = input("Ingrese su Nombre, Direccion, Telefono, Correo, Cuil. Para salir escriba 'salir'.")            
        
        if datos_cliente == "salir":
            condicion = False
        else:
            datos = datos_cliente.split(',')
            nombre = datos[0]
            direccion = datos[1]
            telefono = datos[2]
            correo = datos[3]
            cuil = datos[4]
            
            antiguedad = input("¿Es su primera vez en la empresa? S=Si / N=No")
            
            if antiguedad == "S" or antiguedad == "s":
                pref = False
            elif antiguedad == "N" or antiguedad == "n":
                pref = True
            
            clientes[cuil] = {
                "nombre" : nombre,
                "direccion" : direccion,
                "telefono" : telefono,
                "correo" : correo,
                "preferente" : pref
            } 
    mostrar_clientes()
    
    
def eliminar_cliente():
    condicion = True
    while condicion == True:
        cliente_borrar = input("Ingrese el NIS a borrar. Para salir ingrese 'salir'): ")
        
        if cliente_borrar == "salir":
            condicion = False
        elif cliente_borrar in clientes:
            del clientes[cliente_borrar]
        else:
            "La factura ingresada no existe"
    mostrar_clientes()

def mostrar_cliente():
    condicion = True
    while condicion == True:
        cliente_cuil = input("Ingrese el NIS del cliente. Para salir ingrese 'salir'): ")
        
        if cliente_cuil == "salir":
            condicion = False
        elif cliente_cuil in clientes:
            print("\nDatos del cliente:")
            cliente = clientes[cliente_cuil]
           
            for dato, valor in cliente.items():
                print(dato,":",valor, end=" , ")
            print("CUIL :",cliente_cuil)
        else:
            "El NIS ingresado no existe"

def mostrar_cliente_pref():
    print("\nListado de clientes preferentes:")
    for cuil, datos in clientes.items():
        if datos.get("preferente", False):  
            for dato, valor in datos.items():
                print(dato,":",valor, end=" , ")
            print("CUIL :",cuil)
            

condicion = True
while condicion == True:
    opcion = input("Menú de opciones: (1)Añadir cliente / (2)Eliminar cliente / (3)Mostrar cliente / (4)Listar clientes / (5)Listar clientes preferentes / (6)Terminar. Elige una opcion: ")

    if opcion == "1" or opcion == "(1)":
        añadir_cliente()
    elif opcion == "2" or opcion == "(2)":
        eliminar_cliente()
    elif opcion == "3" or opcion == "(3)":
        mostrar_cliente()
    elif opcion == "4" or opcion == "(4)":
        mostrar_clientes()
    elif opcion == "5" or opcion == "(5)":
        mostrar_cliente_pref()
    elif opcion == "6" or opcion == "(6)":
        condicion = False
    else:
        print("No es un numero válido, vuelva a ingresar otro numero")
```
